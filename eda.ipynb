```python
df_weather = pd.read_json('data/weather_data.json')
df_weather.shape
```

```python
df_weather.head()
```

```python
# Assuming df_weather is your original DataFrame with the 'data' column
# Explode the 'data' column to create new rows for each dictionary
df_weather_exploded = df_weather.explode('data')

# Normalize the exploded data to flatten the nested dictionaries
df_weather_unpacked = pd.json_normalize(df_weather_exploded['data'])

# Optionally, you can merge the unpacked data back with the original DataFrame if needed
df_weather_final = df_weather_exploded.drop(columns=['data']).reset_index(drop=True)
df_weather_final = pd.concat([df_weather_final, df_weather_unpacked.reset_index(drop=True)], axis=1)

# Display the final DataFrame
df_weather_final.shape
```

```python
# Rename the columns by replacing dots with underscores
df_weather_final = df_weather_final.rename(columns={
    'weather.icon': 'weather_icon',
    'weather.code': 'weather_code',
    'weather.description': 'weather_description'
})

df_weather_final.head(5)
```

```python
df_weather_final.to_csv('data/df_weather_final.csv', index=False)
```
